### Removing Duplicate Columns

```python
# The raw data files can contain a duplicate of a column, so it must be removed.
# These duplicate columns contain an integer in their name (e.g "Pharmacy.1")

def remove_dupe_cols(df):
    """
    Returns a df containing columns containing no digits (duplicates)
    """
    cols = set(df.columns)
    dupe_cols = []
    for c in cols:
        if any(char.isdigit() for char in c):
            dupe_cols.append(c)
    # print("Duplicate Columns:")
    # print(dupe_cols)
    df = df.drop(columns=dupe_cols, axis = 1)
    return df

# Example usage:
print("Columns before: " + str(df.shape[1]))
new_df = remove_dupe_cols(df)
print("Columns after: " + str(new_df.shape[1]))
```

### Date Formatting

```python
# Currently, raw data contains 'ts' column with a yyyy-mm-dd-hh-mm-ss format.
# Need to format the 'ts' column, so we get the ddmmyyyy format

def sep_date_and_time(df):
    """
    Returns a df containing two new columns: 'Date' & 'Time' (From the 'ts' column)
    """
    date = []
    time = []

    n = df.shape[0]
    for i in range(n):
        timestamp = df.iloc[i]['ts']
        year = timestamp[0:4]
        month = timestamp[5:7]
        day = timestamp[8:10]

        date.append(day + month + year)
        time.append(timestamp[timestamp.find('T') + 1 : timestamp.find('Z')])

    df.insert(loc = 1, column = "Date", value = date)
    df.insert(loc = 2, column = "Time", value = time)

    return df
```

### Ensuring Float Type

```python
# Convert energy levels into float type in raw dataframe
def convert_to_float(df):
    """
    Returns a df with building columns containing float type values
    """
    ignored_cols = ['ts', 'Date', "Time"]
    building_cols = [c for c in df.columns if c not in ignored_cols]

    df[building_cols] = (
        df[building_cols]
        .replace(r"[^\dEe\+\-\.]", "", regex = True)
        .replace("", np.nan)
        .astype(float)
    )

    return df
```

### Cleaning Data (Putting Everything Together)

```python
def clean(df):
    """
    Returns a df with building columns containing float type values
    """
    return convert_to_float(sep_date_and_time(remove_dupe_cols(df)))
```

### Storing Cleaned Dataframes in 'clean' Folder

```python
def clean_data(filepath):
    """
    Cleans and export the df into a csv file in the clean data directory
    """
    new_filepath = filepath.replace('raw', 'clean')
    # print(filepath)
    raw_df = pd.read_csv(filepath, low_memory=False)
    clean_df = clean(raw_df)
    clean_df.to_csv(new_filepath, index = False)
```

```python
# Collecting the filepaths of all raw skyspark data
def collect_filepaths():
    """
    Return list of filepaths raw data directory
    """
    filepaths = []
    for f in os.listdir('../data/raw/electrical_energy'):
        filepaths.append('../data/raw/electrical_energy/' + f)

    for f in os.listdir('../data/raw/hot_water_energy'):
        filepaths.append('../data/raw/hot_water_energy/' + f)
        
    return filepaths
```

```python
# Running this block will began the cleaning process
for filepath in collect_filepaths():
    clean_data(filepath)
```
